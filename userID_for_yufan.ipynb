```python
import pickle
import gc
import time as tm
import pandas as pd
import os
import matplotlib.pyplot as plt
import re

file_path = "./baidu_contest/exp/"
o_f_name = "userInfo"
dirs2 = sorted(os.listdir(file_path))
user_dic = {}
nFile = 0 
for file in dirs2:
    nFile+=1
    s_time = tm.time()
    fileExt = file.split(".")
    if fileExt[1] == "txt":
        file_name = file_path + file
        fp = open(file_name)
        lines = fp.readlines()
        total_visitor=len(lines)
        fp.close()
        
        for line in lines:
            areaID = file.split("_")[0].split(".")[0]
            classID = int(file.split("_")[-1].split(".")[0])
            data = line.strip('\n').split()
            userID = data[0]
            if not userID in user_dic:
                user_dic[userID] = {} 
            if not areaID in user_dic[userID]:
                user_dic[userID][areaID] = [classID]
            if len(data)> 1:
                timeStamp = data[1].split(",")
                time = {} 
                for each_stamp in timeStamp:
                    time_pieces = each_stamp.split("&")
                    date = time_pieces[0][0:4]+"-"+time_pieces[0][4:6]+ "-" + time_pieces[0][6:8]
                    day = pd.Timestamp(date).day_name()
                    #O(1)~O(n)
                    if not day in time:
                        time[day] = 1 
                    #if not day in time:
                    #    time[day] = {} 
                    #hours = time_pieces[1].split("|")
                    #for each_hour in hours:
                    #    if not each_hour in time[day]:
                    #        time[day][each_hour] = 1
                    #    else:
                    #        time[day][each_hour]+=1
                    else:
                        hours = time_pieces[1].split("|")
                        time[day]+=len(hours)
                if len(user_dic[userID][areaID]) <= 1:
                    user_dic[userID][areaID].append(time)
                else:
                    for each_day in time.keys():
                        if not each_day in user_dic[userID][areaID][1]:
                            user_dic[userID][areaID][1] = {each_day: time[each_day]} 
                        else:
                            user_dic[userID][areaID][1][each_day] += time[each_day]
    #print("before gc #" + str(nFile) + ": " + str(tm.time()-s_time))
    gc.collect()
    #print("after gc #" + str(nFile) + ": " + str(tm.time()-s_time))
    print("#" + str(nFile)+ "/" + str(len(dirs2)) + " finished")
o_file = open(o_f_name, "wb")
pickle.dump(user_dic, o_file)
o_file.close()
```

```python

i_file = open(o_f_name, "rb")
new_dic = pickle.load(i_file)
i_file.close()
#print(new_dic)
dic_area_time = {}
for each_user in new_dic:
    if not each_user in dic_area_time:
        dic_area_time[each_user] = {}
    for each_area in new_dic[each_user]:
        classID = new_dic[each_user][each_area][0]
        if len(new_dic[each_user][each_area]) > 1:
            for each_time in new_dic[each_user][each_area][1]:
                if not each_time in dic_area_time[each_user]:
                    dic_area_time[each_user][each_time] = {each_area: [classID,new_dic[each_user][each_area][1][each_time]]}
                else:
                    if not each_area in dic_area_time[each_user][each_time]:
                        dic_area_time[each_user][each_time][each_area] = [classID, new_dic[each_user][each_area][1][each_time]]
                    else:
                        dic_area_time[each_user][each_time][each_area][1]+= new_dic[each_user][each_area][1][each_time]

o_file = open(o_f_name+'_assorted', "wb") 
pickle.dump(dic_area_time, o_file)
o_file.close()
```

```python
#check repetitive users in test file

#load all users in train dataset
file_path_train = "./baidu_contest/train/"
file_path_test = "./baidu_contest/test/"
dirs_train = sorted(os.listdir(file_path_train))
dirs_test = sorted(os.listdir(file_path_test))
user_dic_train = {}
user_dic_test_notin = {}
user_dic_test_in = {}
#read all userIDs in train dataset
for file in dirs_train:
    fileExt = file.split(".")
    if fileExt[1] == "txt":
        file_name = file_path_train + file
        fp = open(file_name)
        lines = fp.readlines()
        total_visitor=len(lines)
        fp.close()
        
        for line in lines:
            data = line.strip('\n').split()
            userID = data[0]
            if not userID in user_dic_train:
                user_dic_train[userID] = 1
            else:
                user_dic_train[userID]+=1
#read all userIDs in test dataset
for file in dirs_test:
    fileExt = file.split(".")
    if fileExt[1] == "txt":
        file_name = file_path_test + file
        fp = open(file_name)
        lines = fp.readlines()
        total_visitor=len(lines)
        fp.close()
        
        for line in lines:
            data = line.strip('\n').split()
            userID = data[0]
            if not userID in user_dic_train:
                if not userID in user_dic_test_notin:
                    user_dic_test_notin[userID] = 1
                else:
                    user_dic_test_notin[userID] += 1
            else:
                if not userID in user_dic_test_in:
                    user_dic_test_in[userID] = 1
                else:
                    user_dic_test_in[userID]+=1

print("test in train " + str(len(user_dic_test_in)))
print("test not in train " + str(len(user_dic_test_notin)))

o_f_name = "users_in_train"
o_file = open(o_f_name, "wb") 
pickle.dump(user_dic_test_in, o_file)
o_file.close()

```

```python
import pickle
import gc
import time as tm
import pandas as pd
import os
import matplotlib.pyplot as plt
import re

file_path = "./baidu_contest/exp2/"
o_f_name = "userInfoNew"
dirs2 = sorted(os.listdir(file_path))
user_dic = {}
nFile = 0 
for file in dirs2:
    nFile+=1
    s_time = tm.time()
    fileExt = file.split(".")
    if fileExt[1] == "txt":
        file_name = file_path + file
        fp = open(file_name)
        lines = fp.readlines()
        total_visitor=len(lines)
        fp.close()
        
        for line in lines:
            areaID = file.split("_")[0].split(".")[0]
            classID = int(file.split("_")[-1].split(".")[0])
            data = line.strip('\n').split()
            userID = data[0]
            if not userID in user_dic:
                user_dic[userID] = {} 
            if len(data)> 1:
                timeStamp = data[1].split(",")
                for each_stamp in timeStamp:
                    time_pieces = each_stamp.split("&")
                    date = time_pieces[0][0:4]+"-"+time_pieces[0][4:6]+ "-" + time_pieces[0][6:8]
                    day = pd.Timestamp(date).day_name()
                    #O(1)~O(n)
                    if not day in user_dic[userID]:
                        user_dic[userID][day] = {} 
                    hours = time_pieces[1].split("|")
                    for each_hour in hours:
                        if not each_hour in user_dic[userID][day]:
                            user_dic[userID][day][each_hour] = {} 
                        
                        if not classID in user_dic[userID][day][each_hour]:
                            user_dic[userID][day][each_hour][classID] = 1
                        else:
                            user_dic[userID][day][each_hour][classID]+=1
    gc.collect()
    print("#" + str(nFile)+ "/" + str(len(dirs2)) + " finished")
#cast to pandas dataframe
data = pd.DataFrame.from_dict(user_dic, orient = 'index')
o_file = open(o_f_name, "wb")
pickle.dump(data, o_file)
o_file.close()
```
